```python
import pandas as pd
import seaborn as sns

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga':'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=(cod_compostos))
ax.set_title('Top 5 Compostos', fontsize=20)
plt.xlabel('Compostos', fontsize=16)
plt.ylabel('Frequência', fontsize=16)

for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.19, p.get_height()+4))

plt.show()
```

### Desafio 05

```python
fig, axes = plt.subplots(2,2,sharey=True, figsize=(20,8))
g0 = sns.histplot (x='g-0', data=dados, ax=axes[0,0])
g0.set_title('g-0', weight='bold', fontsize=20)
g0.set(xlabel=None) 

g1 = sns.histplot (x='g-1', data=dados, ax=axes[0,1])
g1.set_title('g-1', weight='bold', fontsize=20)
g1.set(xlabel=None) 

g2 = sns.histplot (x='g-2', data=dados, ax=axes[1,1])
g2.set_title('g-2', weight='bold', fontsize=20)
g2.set(xlabel=None) 

g3 = sns.histplot (x='g-3', data=dados, ax=axes[1,0])
g3.set_title('g-3', weight='bold', fontsize=20)
g3.set(xlabel=None) 

fig.tight_layout()
```
